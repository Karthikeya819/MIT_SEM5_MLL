```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/q2_data.csv')
```

```python
df
```

```python
def Naive_Bayes(df, pred_col:str, A:str, B):
    prior_probability = 1
    evidence = 1
    for feature in B:
        total_count = df[feature].count()
        count = ((df[pred_col] == A) & (df[feature] == B[feature])).astype(int).sum()
        prior_probability *= float(count) / (df[pred_col] == A).astype(int).sum()
        evidence *= float((df[feature] == B[feature]).astype(int).sum()) / total_count
    
    likelihood = float((df[pred_col] == A).astype(int).sum()) / df[pred_col].count()

    return (prior_probability * likelihood)/evidence
```

```python
B = {
    'age': '<=30',
    'income': 'medium',
    'student': 'yes',
    'credit_rating': 'fair'
}
pred_col = 'buys_computer'
probs = {}
for A in df[pred_col]:
    probs[A] = Naive_Bayes(df, 'buys_computer', A, B)

print(f'Prediction for given conditions is {max(probs)}')
```
