1. Follow along with these steps: <br>
    a) Create a figure object called fig using plt.figure()<br>
    b) Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.<br>
    c) Plot (x,y) on that axes and set the labels and titles to match the plot below:<br>

```python
x = [0, 25, 50, 100]
y = [0, 50, 100, 200]

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

plt.show()
```

4. Use the company sales dataset csv file, get total profit of all months and show line plot with the following<br>
   Style properties. Generated line plot must include following Style properties: –<br>
    a. Line Style dotted and Line-color should be red<br>
    b. Show legend at the lower right location.<br>
    c. X label name = Month Number<br>
    d. Y label name = Sold units number<br>
    e. Add a circle marker.<br>
    f. Line marker color as read<br>
    g. Line width should be 3 <br>

```python
df = pd.read_csv('company_sales_data.csv')
x, y = df['month_number'], df['total_profit']

plt.plot(x, y, linestyle=':', color= 'red', marker='o', label='Sold units', markerfacecolor='red', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.title('Company Profit Analysis')
plt.legend(loc='lower right')

plt.show()
```

A1. Use the company sales dataset csv file, read all product sales data and show it  using a multiline plot.Display the number of units sold per month or each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
df = pd.read_csv('company_sales_data.csv')
x = df['month_number']
products = df.columns[1:-2]

for product in products:
    y = df[product]
    plt.plot(x, y, label=product)

plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.legend()

plt.show()
```

A2. Use the company sales dataset csv file, calculate total sale data for last year for each product and show it using a Pie chart. Note: In Pie chart display Number of units sold per year for each product in percentage.

```python
df = pd.read_csv('company_sales_data.csv')
products = df.columns[1:-2]

ind_units = df.sum()[products]
plt.pie(ind_units, labels=products, autopct='%1.1f%%')

plt.show()
```
